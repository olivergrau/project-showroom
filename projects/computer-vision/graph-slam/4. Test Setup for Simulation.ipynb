## Experiments with graph_slam

In this notebook I set up a test routine and I extended the existing make_data function to return the true path along the data. My `run_tests` function takes in one or multiple parameter sets, produces data for the parameter set and executes the slam function. It then draws the real and the estimated path and calculates the distance between the robots real last pose and the estimated pose.

```python
%load_ext autoreload
%autoreload 2

import numpy as np
import matplotlib.pyplot as plt
import random
from math import pi, cos, sin
from robot_class import robot
from slam import slam

# Assuming the `robot` class and `slam` function are already defined as per your code.

from helpers import make_data

# --- Test routine ---
def run_tests(parameter_sets):
    results = []

    for idx, params in enumerate(parameter_sets):
        print(f"\nRunning test {idx + 1} with parameters:")
        print(params)
        N = params['N']
        num_landmarks = params['num_landmarks']
        world_size = params['world_size']
        measurement_range = params['measurement_range']
        motion_noise = params['motion_noise']
        measurement_noise = params['measurement_noise']
        distance = params['distance']

        # Generate data and true path
        data, true_path = make_data(N, num_landmarks, world_size, measurement_range,
                                    motion_noise, measurement_noise, distance)

        # Run SLAM
        mu = slam(data, N, num_landmarks, world_size, motion_noise, measurement_noise)

        # Extract estimated path from mu
        estimated_path = []
        for i in range(N):
            x_est = mu[2 * i, 0]
            y_est = mu[2 * i + 1, 0]
            estimated_path.append([x_est, y_est])

        # Calculate the Euclidean distance between the true and estimated final poses
        true_final_pose = np.array(true_path[-1])
        est_final_pose = np.array(estimated_path[-1])
        difference = est_final_pose - true_final_pose
        distance_error = np.linalg.norm(difference)
        print(f"Distance between estimated and true final poses: {distance_error}")

        # Store the results
        results.append({
            'true_path': true_path,
            'estimated_path': estimated_path,
            'distance_error': distance_error,
            'params': params
        })

        # Plot the paths
        plt.figure(figsize=(8, 6))
        true_path_np = np.array(true_path)
        estimated_path_np = np.array(estimated_path)
        plt.plot(true_path_np[:, 0], true_path_np[:, 1], 'b-', label='True Path')
        plt.plot(estimated_path_np[:, 0], estimated_path_np[:, 1], 'r--', label='Estimated Path')
        plt.title(f"Test {idx + 1} - True vs Estimated Path")
        plt.xlabel('X')
        plt.ylabel('Y')
        plt.legend()
        plt.grid(True)
        plt.show()

    return results
```

```python
# --- Example parameter sets ---
parameter_sets = [
    {
        'N': 20,
        'num_landmarks': 5,
        'world_size': 100.0,
        'measurement_range': 50.0,
        'motion_noise': 1.0,
        'measurement_noise': 1.0,
        'distance': 20.0
    },
    {
        'N': 50,
        'num_landmarks': 10,
        'world_size': 200.0,
        'measurement_range': 100.0,
        'motion_noise': 0.5,
        'measurement_noise': 0.5,
        'distance': 10.0
    },
    {
        'N': 30,
        'num_landmarks': 8,
        'world_size': 150.0,
        'measurement_range': 75.0,
        'motion_noise': 2.0,
        'measurement_noise': 2.0,
        'distance': 15.0
    },
    {
        'N': 30,
        'num_landmarks': 3,
        'world_size': 20,
        'measurement_range': 10.0,
        'motion_noise': 2.0,
        'measurement_noise': 2.0,
        'distance': 5.0
    },
    {
        'N': 30,
        'num_landmarks': 8,
        'world_size': 150.0,
        'measurement_range': 75.0,
        'motion_noise': 10.0,
        'measurement_noise': 10.0,
        'distance': 15.0
    },
]

# --- Run the tests ---
results = run_tests(parameter_sets)
```

## Interpretation of the results:

We can clearly see here, that the world size has a siginficant impact on the true and estimated pathes. As less information slam has the less accurate are the results. And this even with small noise levels. With a larger world (>=100) the results are more accurate if the noise levels are reasonable but when I increase noise levels again (for example 10 or more) than even setups with large world sizes have signififant noise and differences between real and estimated path. 

This is to be expected.
